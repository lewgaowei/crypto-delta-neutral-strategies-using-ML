```python
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------
# List your four CSV file paths here
# ----------------------------------------------------
csv_files = [
    "D:\\Homework\\QF634\\project\\Data\\raw_funding_rate\\LTCUSDT_funding_rate_20200101_20251130.csv",
]

for csv in csv_files:
    # Load CSV
    df = pd.read_csv(csv)

    # Expect columns: Time, Value
    # If yours are named differently, change below
    time_col = df.columns[4]
    value_col = df.columns[1]

    # ---------------------------------------------
    # 1. Time–Value Line Chart (no y-label)
    # ---------------------------------------------
    plt.figure(figsize=(8,3))
    plt.plot(df[time_col], df[value_col])
    plt.title(f"Time–Value Chart: {csv}")
    plt.xlabel("Time")        # no ylabel
    plt.tight_layout()
    plt.show()

    # ---------------------------------------------
    # 2. Distribution in Log Scale
    # ---------------------------------------------
    plt.figure(figsize=(4,4))
    plt.hist(df[value_col], bins=40)
    plt.yscale("log")         # log-scale
    plt.title(f"Distribution (Log Scale): {csv}")
    plt.xlabel("Value")       # auto Y log scale
    plt.tight_layout()
    plt.show()
```
